```python
import numpy as np
import matplotlib.pyplot as plt
import h5py

def averaged_profiles(profile):
    """Return average profile by normalizing by count and filtering"""
    out_var = profile[(profile[:, 1] > 1)]  # remove rows with no data
    out_var = np.array([out_var[:, 0] / out_var[:, 1], out_var[:, 2]]).T
    return out_var

def get_data(type_, var_list, dir, res, t_res, Np_str):
    """Load profiles for given type and variables from HDF5 files"""
    input_file = f"{dir}/Project_Algorithms/Domain_Profiles/OUTPUT/{type_}_eulerian_buoyancy_profiles_{res}_{t_res}_{Np_str}.h5"
    data_dict = {}
    with h5py.File(input_file, 'r') as f:
        for var in var_list:
            data_dict[f"profile_{var}"] = np.array(f[f"profile_{var}"])
    return data_dict

def average_difference(array1, array2):
    """Compute difference between averaged profiles where z matches"""
    out_var_one = averaged_profiles(array1)
    out_var_two = averaged_profiles(array2)

    # mask to keep matching heights
    mask_one = np.isin(out_var_one[:, 1], out_var_two[:, 1])
    mask_two = np.isin(out_var_two[:, 1], out_var_one[:, 1])

    out_var_one = out_var_one[mask_one]
    out_var_two = out_var_two[mask_two]

    diff = out_var_one[:, 0] - out_var_two[:, 0]
    zs = out_var_one[:, 1]

    return np.column_stack((diff, zs))

def diff_plotting(vars, xlabels, title, dir, res, t_res, Np_str, cloudbase, mean_LFC):
    """Return cloudy minus general updraft difference plots"""
    n_vars = len(vars)
    n_cols = int(np.ceil(np.sqrt(n_vars)))
    n_rows = int(np.ceil(n_vars / n_cols))
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows))
    fig.suptitle(title)

    axs = np.array([axs]) if n_vars == 1 else axs.flatten()

    # Load cloudy/general data once each
    data_cloudy = get_data("cloudy", vars, dir, res, t_res, Np_str)
    data_general = get_data("general", vars, dir, res, t_res, Np_str)

    for var, axis, xlabel in zip(vars, axs, xlabels):
        profile_cloudy = data_cloudy[f"profile_{var}"]
        profile_general = data_general[f"profile_{var}"]

        diff_profile = average_difference(profile_cloudy, profile_general)

        if var in ['qv', 'qc_plus_qi', 'qi', 'HMC']:
            diff_profile[:, 0] *= 1000  # convert to g/kg or g/kg/s

        axis.plot(diff_profile[:, 0], diff_profile[:, 1], color='black')
        axis.axvline(0, color='black', linestyle='dashed')

        axis.set_ylabel('z (km)')
        axis.set_xlabel(xlabel)
        axis.grid(True)
        axis.axhline(cloudbase, color='purple', linestyle='dashed')
        axis.axhline(mean_LFC / 1000, color='green', linestyle='dashed')

        # Apply scientific notation if you have this defined
        if 'apply_scientific_notation' in globals():
            apply_scientific_notation([axis])

    #hiding leftover axises
    for ax in axs[n_vars:]:
        ax.set_visible(False)

    #coordinating xlimits
    fix_x_limits(axs);
    fix_y_limits(axs);

    plt.tight_layout()
    # plt.savefig(f"{dir}/Project_Algorithms/plots/eulerian_updraft_profiles_diff_{res}_{t_res}_{Np_str}.jpg", bbox_inches='tight', dpi=300)
```

```python
vars = ['buoyancy', 'th_term', 'rv_term', 'rl_term']
xlabels = [
    r"Buoyancy (m/s²)",
    r"$\theta'$ component (m/s²)",
    r"$r_v'$ component (m/s²)",
    r"$r_l'$ component (m/s²)",
]
title = "Cloudy - General Updraft Updraft Property Profiles (Eulerian)"

diff_plotting(
    vars=vars,
    xlabels=xlabels,
    title=title,
    dir=dir,
    res=res,
    t_res=t_res,
    Np_str=Np_str,
    cloudbase=cloudbase,
    mean_LFC=mean_LFC
)
```
