```python
def plot_mean_entrainment(ax, profile_array_e, profile_array_d, title):
    zh=data['zh'].data

    # Compute mean profiles
    e = np.mean(profile_array_e, axis=0)
    d = np.mean(profile_array_d, axis=0)
    net = np.mean(profile_array_e - profile_array_d, axis=0)

    # Plot
    ax.plot(e, zh, color='blue', label='Entrainment')
    ax.plot(d, zh, color='red', label='Detrainment')
    ax.plot(net, zh, color='black', label='Net Entrainment')
    ax.axvline(0, color='black')

    ax.axhline(cloudbase, color='purple', linestyle='dashed', label='Cloud base',lw=1.2)
    ax.axhline(MeanLFC / 1000, color='green', linestyle='dashed', label='Mean LFC',lw=1.2)
    
    ax.set_title(f"{title}")
    ax.set_xlabel(r"($kg m^{-3} s^{-1}$)")  
    ax.set_ylabel('z (km)')
    ax.legend()

    # Format x-axis in scientific notation
    apply_scientific_notation([ax])
```

```python
def plot_transfer_rate(ax, c_to_g, g_to_c,title):
    zh=data['zh'].data

    # Mean profiles
    c_to_g_mean = np.mean(c_to_g, axis=0)
    g_to_c_mean = np.mean(g_to_c, axis=0)

    # Plot transfers
    ax.plot(c_to_g_mean, zh, color='forestgreen', label='Cloudy → General')
    ax.plot(g_to_c_mean, zh, color='dodgerblue', label='General → Cloudy')
    ax.axvline(0, color='black', linewidth=1)

    # Reference lines
    ax.axhline(cloudbase, color='purple', linestyle='dashed',lw=1.2)
    ax.axhline(MeanLFC / 1000, color='forestgreen', linestyle='dashed',lw=1.2)

    # Labeling and formatting
    ax.set_title(f"{title}")
    ax.set_xlabel('Mass Transfer Rate')
    ax.set_xlabel(r"($kg m^{-3} s^{-1}$)")  
    ax.set_ylabel('z (km)')
    ax.set_xlim(left=0)
    ax.legend()

    apply_scientific_notation([ax])

def plot_transfer_ratio(ax, profile_array_e_g, c_to_g, profile_array_e_c, g_to_c, title):
    zh = data['zh'].data

    # Compute mean profiles first
    mean_c_to_g = np.mean(c_to_g, axis=0)
    mean_g_to_c = np.mean(g_to_c, axis=0)
    mean_e_g = np.mean(profile_array_e_g, axis=0)
    mean_e_c = np.mean(profile_array_e_c, axis=0)

    # Mask ratios where denominator is too small
    threshold = 1e-7
    with np.errstate(divide='ignore', invalid='ignore'):
        ratio_c_to_g = np.where(mean_e_g > threshold, mean_c_to_g / mean_e_g, np.nan)
        ratio_g_to_c = np.where(mean_e_c > threshold, mean_g_to_c / mean_e_c, np.nan)

    # Plot ratios
    ax.plot(ratio_c_to_g, zh, color='blue', label='Cloudy → General / General Ratio')
    ax.plot(ratio_g_to_c, zh, color='red', label='General → Cloudy / Cloudy Ratio')
    ax.axvline(1, color='black', linestyle='dashed', linewidth=1)

    # Reference horizontal lines
    ax.axhline(cloudbase, color='purple', linestyle='dashed', lw=1.2)
    ax.axhline(MeanLFC / 1000, color='green', linestyle='dashed', lw=1.2)

    # Labels and limits
    ax.set_title(f"{title}")
    ax.set_xlabel('Ratio')
    ax.set_ylabel('z (km)')
    pad_fraction = 10
    pad_multiplier = (100 + pad_fraction) / 100
    ax.set_xlim(0, 1 * pad_multiplier)

    ax.legend()
    apply_scientific_notation([ax])
```
