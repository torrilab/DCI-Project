```python
import numpy as np
```

```python
D=np.zeros((4,4),dtype=int);np.fill_diagonal(D, 1)
X=np.zeros_like(D); X[:] = np.arange(1, 5).reshape(-1, 1)
print(f'D={D}\n');print(f'X={X}\n')

Nx=5
output = np.zeros((Nx))

np.add.at(output,(X),D)
print(f'output={output}')
```

```python
D=np.zeros((4,4),dtype=int);np.fill_diagonal(D, 1);
X=np.zeros_like(D); X[:] = np.arange(1, 5).reshape(-1, 1)
Y=np.zeros_like(D); Y[:] = np.arange(1, 5).reshape(-1, 1)
print(f'D={D}\n');print(f'X={X}\n')

Ny=5;Nx=5
output = np.zeros((Ny, Nx))

np.add.at(output,(Y,X),D)
print(f'output={output}')
```

```python
D=np.zeros((4,4),dtype=int);np.fill_diagonal(D, 1);
X=np.zeros_like(D); X[:] = np.arange(1, 5).reshape(-1, 1)
Y=np.zeros_like(D); Y[:] = np.arange(1, 5).reshape(-1, 1)
Z=np.zeros_like(D); Z[:] = np.arange(1, 5).reshape(-1, 1)
print(f'D={D}\n');print(f'X={X}\n')

Nz=5; Ny=5;Nx=5
output = np.zeros((Nz, Ny, Nx))

np.add.at(output,(Z,Y,X),D)
print(f'output={output}')
```
